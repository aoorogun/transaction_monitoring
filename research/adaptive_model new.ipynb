```python
import pandas as pd
import polars as pl
import numpy as np
from river import drift
from river import stream
from river import tree
from river import metrics
from river import imblearn
from river import evaluate
from river import neural_net as nn
from river import facto
from river import preprocessing as pp
from river import optim
from sklearn.neural_network import MLPClassifier
```

```python

big_df = pd.read_csv("big_data_new.csv")

```

```python
big_df.columns
```

```python
big_df = big_df.drop(['ssn', 'cc_num', 'first', 'last', 'street', 'city', 'state','acct_num',
       'zip', 'lat', 'long', 'city_pop', 'merchant', 'merch_lat', 'merch_long', 'profile',
       'trans_num', 'trans_date', 'trans_time', 'unix_time', 'category','job'], axis = 1)
```

```python
pdf = big_df.copy()
```

```python


hatc = tree.HoeffdingAdaptiveTreeClassifier(
    grace_period=100,
    delta=1e-5,
    leaf_prediction='nb',
    nb_threshold=10,
    seed=0
)

hatc_eddm = tree.HoeffdingAdaptiveTreeClassifier(
    grace_period=100,
    delta=1e-6,
    leaf_prediction='nb',
    nb_threshold=10,
    seed=0,
    drift_detector=drift.binary.EDDM()
)

hatc_ddm = tree.HoeffdingAdaptiveTreeClassifier(
    grace_period=100,
    delta=1e-5,
    leaf_prediction='nb',
    nb_threshold=10,
    seed=0,
    drift_detector=drift.binary.DDM()
)

hatc_eddm2 = tree.HoeffdingAdaptiveTreeClassifier(
    grace_period=100,
    delta=1e-5,
    leaf_prediction='nb',
    nb_threshold=10,
    seed=0,
    drift_detector=drift.binary.EDDM()
)

drc_hatc = drift.DriftRetrainingClassifier(
    model=tree.HoeffdingAdaptiveTreeClassifier(
    grace_period=100,
    delta=1e-5,
    leaf_prediction='nb',
    nb_threshold=10,
    seed=0),
    drift_detector=drift.binary.EDDM()
)

hsc_hatc = imblearn.HardSamplingClassifier(
        classifier= tree.HoeffdingAdaptiveTreeClassifier(
            grace_period=100,
            delta=1e-5,
            leaf_prediction='nb',
            nb_threshold=10,
            seed=0
        ),
        p=0.1,
        size=40,
        seed=42,
    )

mlp = (
    pp.StandardScaler() |
    nn.MLPRegressor(
        hidden_dims=(5,),
        activations=(
            nn.activations.ReLU,
            nn.activations.ReLU,
            nn.activations.Identity
        ),
        optimizer=optim.SGD(1e-3),
        seed=42
    )
)

mlp_online = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1, random_state=42)


ffm = facto.FFMClassifier(
    n_factors=10,
    intercept=.5,
    seed=42,
)
```

```python
metric = metrics.Accuracy()
report1 = metrics.ClassificationReport()
report2 = metrics.ClassificationReport()
report3 = metrics.ClassificationReport()
report4 = metrics.ClassificationReport()
report5 = metrics.ClassificationReport()
report6 = metrics.ClassificationReport()
report7 = metrics.ClassificationReport()
report8 = metrics.ClassificationReport()


roc1 = metrics.ROCAUC()
roc2 = metrics.ROCAUC()
roc3 = metrics.ROCAUC()
roc4 = metrics.ROCAUC()
roc5 = metrics.ROCAUC()
roc6 = metrics.ROCAUC()
roc7 = metrics.ROCAUC()
roc8 = metrics.ROCAUC()
```

```python
models = ["HATC", "HATC-EDDM", "HATC-DDM", "HATC-EDDM2", "DRC-HATC", "HSC-HATC", "NN", "FFM"]
```

```python
i = 0
n = 0
last_idx = 0
current_idx = 0
drift_detector1 = drift.binary.DDM()
drift_detector2 = drift.binary.DDM()
drift_detector3 = drift.binary.DDM()
drift_detector4 = drift.binary.DDM()
drift_detector5 = drift.binary.DDM()
drift_detector6 = drift.binary.DDM()
drift_detector7 = drift.binary.DDM()
drift_detector8 = drift.binary.DDM()

drifts = []
y_pred = []
t_all = []
HATC = []
HATC_EDDM = []
HATC_DDM = []
HATC_EDDM2 = []
DRC_HATC = []
HSC_HATC = []
NN = []
FFM = []
y_orig = []

for x, t in stream.iter_pandas(pdf.drop('is_fraud', axis=1), pdf.is_fraud):

    hatc.learn_one(x, t)
    hatc_eddm.learn_one(x, t)
    hatc_ddm.learn_one(x, t)
    hatc_eddm2.learn_one(x, t)
    drc_hatc.learn_one(x, t)
    hsc_hatc.learn_one(x, t)
    # nn.learn_one(x, t)
    mlp_online.partial_fit(np.array(list(x.values())).reshape(1,-1), np.array([t]), classes=np.array([0,1]))

    ffm.learn_one(x, t)

    i += 1
    # x = pl.DataFrame(x).to_pandas()
    # t = pd.DataFrame({"t":t}, index=[0])
    y_pred1 = hatc.predict_one(x)
    y_pred2 = hatc_eddm.predict_one(x)
    y_pred3 = hatc_ddm.predict_one(x)
    y_pred4 = hatc_eddm2.predict_one(x)
    y_pred5 = drc_hatc.predict_one(x)
    y_pred6 = hsc_hatc.predict_one(x)
    y_pred7 = mlp_online.predict(np.array(list(x.values())).reshape(1,-1)).item()
    y_pred8 = ffm.predict_one(x)

    #append all scores to be saved
    HATC.append(y_pred1)
    HATC_EDDM.append(y_pred2)
    HATC_DDM.append(y_pred3)
    HATC_EDDM2.append(y_pred4)
    DRC_HATC.append(y_pred5)
    HSC_HATC.append(y_pred6)
    NN.append(y_pred7)
    FFM.append(y_pred8)
    y_orig.append(t)

    roc1.update(t, hatc.predict_proba_one(x))
    roc2.update(t, hatc_eddm.predict_proba_one(x))
    roc3.update(t, hatc_ddm.predict_proba_one(x))
    roc4.update(t, hatc_eddm2.predict_proba_one(x))
    roc5.update(t, drc_hatc.predict_proba_one(x))
    roc6.update(t, hsc_hatc.predict_proba_one(x))
    probs = mlp_online.predict_proba(np.array(list(x.values())).reshape(1,-1))
    roc7.update(t, {i: float(val) for i, val in enumerate(probs[0])})
    roc8.update(t, ffm.predict_proba_one(x))
    #y_pred = check_y(y_pred)
    # print(y_pred, type(y_pred))
    #metric = metric.update(t, y_pred)
    report1.update(t, y_pred1)
    report2.update(t, y_pred2)
    report3.update(t, y_pred3)
    report4.update(t, y_pred4)
    report5.update(t, y_pred5)
    report6.update(t, y_pred6)
    report7.update(t, y_pred7)
    report8.update(t, y_pred8)

    if i%1_000_000 == 0:
    # if i%10 == 0:
        # print(metric)
        n += 1_000_000
        # n += 10
        print(f"{n} samples")
        reports = [report1, report2, report3, report4, report5, report6, report7, report8]
        rocs = [roc1, roc2, roc3, roc4, roc5, roc6, roc7, roc8]
        for m, r in zip(models, rocs):
            print(m)
            print(r)
            print("++++++++===========++++++++")
        for m1, r1 in zip(models, reports):
            print(m1)
            print(r1)
            print("++++++++===========++++++++")
        print("=================================")

    if y_pred1 is not None:
        drift_detector1.update(y_pred1)
        if drift_detector1.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")
            #print(pdf.drop('is_fraud', axis=1).iloc[last_idx:current_idx])
            print(f"from {last_idx} to {current_idx}")
            print("=========================================")
            # print(dd)
            drifts.append(i)
            drift_detector1 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred2 is not None:
        drift_detector2.update(y_pred2)
        if drift_detector2.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector2 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred3 is not None:
        drift_detector3.update(y_pred2)
        if drift_detector3.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector3 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred4 is not None:
        drift_detector4.update(y_pred2)
        if drift_detector4.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector4 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred5 is not None:
        drift_detector5.update(y_pred2)
        if drift_detector5.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector5 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred6 is not None:
        drift_detector6.update(y_pred2)
        if drift_detector6.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector6 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred7 is not None:
        drift_detector7.update(y_pred2)
        if drift_detector7.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector7 = drift.binary.DDM()
            last_idx = current_idx

    if y_pred8 is not None:
        drift_detector8.update(y_pred2)
        if drift_detector8.drift_detected:
            print(f'Change detected at index {i}')
            current_idx = i
            print("Data that contributed to drift\n\n")
            print("=========================================")

            print(f"from {last_idx} to {current_idx}")
            print("=========================================")

            drift_detector8 = drift.binary.DDM()
            last_idx = current_idx

```
